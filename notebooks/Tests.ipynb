```python
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
```

```python
project_dir = Path().resolve().parents[0]
```

```python
train_X = np.load(project_dir / "data/processed/" / "train_X_2D.npy")
train_y = np.load(project_dir / "data/processed/" / "train_y.npy")
```

```python
train_X_mean = np.mean(train_X[train_y == 9], axis=0)
```

```python
train_X_mean.shape
```

```python
fig, axs = plt.subplots(nrows=2, ncols=5, figsize=(20, 10))

extent = [0, 28, 0, 28]

for row, col in zip([0] * 5 + [1] * 5, range(10)):
    axs[row][col % 5].imshow(np.mean(train_X[train_y == col], axis=0), cmap=plt.get_cmap("coolwarm"), extent=extent)
    axes_range = range(0, 29, 7)

    axs[row][col % 5].set_xticks(axes_range)
    axs[row][col % 5].set_yticks(axes_range)
```
